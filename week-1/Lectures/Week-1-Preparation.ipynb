# Git installation

Git is a very commonly used version control system. When you start working on complicated projects using a reliable version control system is a must. So again, install and start working with git. Sooner is better.

- For Mac and Linux users: You don't have to install git as it comes installed on your computers.

- For Windows users please follow the directions in this video:
